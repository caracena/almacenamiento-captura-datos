### Ejercicio 1

El primer ejercicio consiste en crear un archivo de texto (nombres.txt) con nombres y apellidos. Luego, leer ese archivo y crear un archivo csv (nombres.csv) con 1 columna de nombres y otra de apellidos. 

```python
import names
import csv
```

```python
names.get_full_name()
```

```python
with open('data/nombres.txt', 'w') as f:
    for i in range(100):
        f.write(names.get_full_name() + '\n')
```

```python
with open('data/nombres.txt', 'r') as f1:
    content = f1.readlines()
    with open('data/nombres.csv', 'w') as f2:
        writer = csv.writer(f2)
        for line in content:
            row = line.strip().split(" ")
            writer.writerow(row)
```

### Ejercicio 2

El segundo ejercicio consiste en contar la frecuencia de las palabras de un archivo de texto y mostrar las 10 palabras mas frecuentes.

```python
with open('data/el_quijote.txt', 'r') as f:
    content = f.read()
    
words = content.split(" ")
counter = {}

for w in words:
    if w in counter:
        counter[w] += 1
    else:
        counter[w] = 0

sorted_counter = sorted(counter, key=counter.get, reverse=True)
for w in sorted_counter[:10]:
    print(w, counter[w])
```

### Ejercicio 5

```python
from xml.etree import ElementTree as ET

tree = ET.parse('data/data-text.xml')
root = tree.getroot()
print(root)

all_data = []

for observation in root:
    record = {}
    for item in observation:
        record[item.tag] = item.text
        
    all_data.append(record)

print(all_data)
```

### Ejercicio 6

Importar un archivo excel con libreria xlrd

```python
import xlrd
book = xlrd.open_workbook("data/SOWC 2014 Stat Tables_Table 9.xlsx")

# sheet = book.sheet_by_name("Table 9 ")
sheet = book.sheet_by_index(1)

print(sheet.name)
print(sheet.nrows)
print(sheet.ncols)

data = {}
for i in range(14, sheet.nrows):

    # Start at 14th row, because that is where the countries begin
    row = sheet.row_values(i)

    country = row[1]

    data[country] = {
        'child_labor': {
            'total': [row[4], row[5]],
            'male': [row[6], row[7]],
            'female': [row[8], row[9]],
        },
        'child_marriage': {
            'married_by_15': [row[10], row[11]],
            'married_by_18': [row[12], row[13]],
        }
    }

    if country == "Zimbabwe":
        break

print(data)
```
