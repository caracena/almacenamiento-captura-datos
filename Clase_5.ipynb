Obtener la lista de feriados del 2018 de la API de https://www.feriadosapp.com/api/holidays.json

### Ejercicio 2

Obtener las 10 Ãºltimas leyes chilenas de la API https://www.leychile.cl/Consulta/legislacion_abierta_web_service y crear un dataframe con ellas

```python
import requests
from xml.etree import ElementTree as ET

payload = {
    'opt': 3,
    'cantidad': 10
}

r = requests.get('https://www.leychile.cl/Consulta/obtxml', params=payload)
tree = ET.fromstring(r.text)
```

```python
for n in tree:
    for elem in n:
        print(elem.tag, elem.text)
```

```python
leyes = []
for n in tree:
    norma = {}
    for elem in n:
        if elem.tag in ['FECHA_PUBLICACION', 'FECHA_PROMULGACION', 'TITULO', 'TEXTO']:
            norma[elem.tag] = elem.text
    leyes.append(norma)
```

```python
leyesdf = pd.DataFrame(leyes) 
leyesdf
```

### Ejercicio 5

Usar la librerÃ­a tweepy para leer tweets en streaming con la palabra python en su contenido

```python
import tweepy
#override tweepy.StreamListener to add logic to on_status
class MyStreamListener(tweepy.StreamListener):

    def on_status(self, status):
        print(status.text)
        print()
        
myStreamListener = MyStreamListener()
myStream = tweepy.Stream(auth = api.auth, listener=myStreamListener)

myStream.filter(track=['python'])
```
