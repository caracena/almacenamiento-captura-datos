### Ejercicio 1

Conectarse a una base de datos sqlite y extraer todos los clientes

```python
import sqlite3

conn = sqlite3.connect("data/chinook.db")

cur = conn.cursor()

cur.execute("SELECT * FROM customers")

rows = cur.fetchall()

print(rows)

for row in rows:
    print(row[7])
```

Conectarse a la base de datos y extraer todos los clientes con pandas

### Ejercicio 5

Realizar consultas en pandas sin usar SQL

```python
## SELECT FirstName FROM Customers;

customers = pd.read_sql("SELECT * FROM customers", conn)
customers['FirstName']
```

```python
## SELECT DISTINCT Country FROM Customers;

customers['Country'].unique()
```

```python
customers['Country'].value_counts()
```

```python
## SELECT DISTINCT City, Country FROM Customers;

customers[['Country', 'City']].drop_duplicates()
```

```python
## SELECT * FROM Customers ORDER BY Country;

customers.sort_values('Country')
```

```python
## SELECT * FROM Customers ORDER BY Country DESC;

customers.sort_values('Country', ascending=False)
```

```python
## SELECT FirstName, City, Country FROM Customers ORDER BY Country ASC, City DESC;

customers.sort_values(['Country', 'City'], ascending=[True, False])
```

```python
## SELECT * FROM Customers WHERE Country = 'Brazil';

customers[customers['Country'] == 'Brazil']
```

```python
## SELECT * FROM Customers WHERE Country = 'USA' AND City = 'Mountain View';

customers[(customers['Country'] == 'USA') & (customers['City'] == 'Mountain View')]
```

```python
## SELECT * FROM Customers WHERE Country = 'Chile' OR City = 'Buenos Aires';

customers[(customers['Country'] == 'Chile') | (customers['City'] == 'Buenos Aires')]
```

```python
## SELECT * FROM Customers WHERE Country IN ('Chile', 'Australia');

customers[customers['Country'].isin(['Chile', 'Australia'])]
```

```python
## SELECT * FROM Customers WHERE Country LIKE 'A%';

customers[customers['Country'].str.startswith('A')]
```

```python
## SELECT *, 1.19 * Total FROM 'invoices';

invoices = pd.read_sql("SELECT * FROM invoices", conn)
invoices['Total_new'] = invoices['Total'] * 1.19
invoices
```

```python
## SELECT min(Total) FROM Invoices;
invoices['Total'].min()
```

```python
invoices.min()
```

```python
## SELECT max(Total) FROM Invoices;
invoices['Total'].max()
```

```python
invoices['Total'].describe()[['mean', 'count']]
```

```python
## SELECT sum(Total) FROM Invoices WHERE BillingCountry = 'Germany';
invoices[invoices['BillingCountry'] == 'Germany']['Total'].sum()
invoices.loc[invoices['BillingCountry'] == 'Germany','Total'].sum()
```

```python
## SELECT round(avg(Total),2) FROM Invoices WHERE BillingCountry = 'Germany';

invoices[invoices['BillingCountry'] == 'Germany']['Total'].mean().round(2)
```

```python
# SELECT BillingCountry, round(sum(Total),2) monto, count(Total) cantidad, round(avg(Total),2) promedio 
# FROM Invoices 
# GROUP BY BillingCountry
# ORDER BY monto DESC;

invoices.groupby('BillingCountry')['Total'].agg(['sum','count', 'mean']).sort_values('sum', ascending=False)
```
